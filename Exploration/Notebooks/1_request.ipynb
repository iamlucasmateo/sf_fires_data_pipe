```python
import requests
from config import api_config
```

```python
APP_TOKEN = api_config["app_token"]
SECRET_TOKEN = api_config["secret_token"]
DATASET_ID = "wr8u-xric"
DATA_URL = f"https://data.sfgov.org/resource/{DATASET_ID}.json"
COUNT_QUERY = "?$select=count(ID)"

headers = {
  "X-App-Token": APP_TOKEN, # config.app_token
  "Content-Type": "application/json"
}
```

```python
req_count = requests.get(
    f"{DATA_URL}{COUNT_QUERY}",
    headers=headers
)
total_rows = req_count.json()[0]["count_ID"]

LIMIT_QUERY = f"?$limit={total_rows}"

req = requests.get(
    f"{DATA_URL}{LIMIT_QUERY}",
    headers=headers
)
data = req.json()
```
