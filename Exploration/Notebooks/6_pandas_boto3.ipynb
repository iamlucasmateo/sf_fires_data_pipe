```python
def get_coord(coord_type):
    def get_item_coord(item):
        try:
            latitude, longitude = item["coordinates"]
            if coord_type == "latitude":
                return latitude
            return longitude
        except:
            return None
    return get_item_coord
```

```python
df["latitude"] = df["point"].apply(get_coord("latitude"))
df["longitude"] = df["point"].apply(get_coord("longitude"))
```

```python
df = df.drop(columns=["point"])
```
